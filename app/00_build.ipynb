```python
import pandas as pd
import numpy as np
import os
import bz2 
import random
from random import randint
random.seed(42)
```

# Read Data

```python
fname = 'amazon_train.pk'
fpath = os.path.join(INPUT_DIR, fname)
idx = [random.randint(0, 3600000) for _ in range(200)]

if os.path.isfile(fpath):
    data = pd.read_pickle(fpath)
    data = data.iloc[idx]

else:
    zipfname = 'train.ft.txt.bz2'
    zippath = os.path.join(INPUT_DIR, zipfname)
    data = bz2.BZ2File(zippath)
    data = [x.decode('utf-8') for x in data.readlines()]
    data = pd.DataFrame(data)
    data.to_pickle(fpath)
    data = data.iloc[idx]
    data.columns = ['text']

data.sample()
```

```python
df = data.copy().reset_index()
df.columns = ['id', 'text']
df[['label', 'text']] = df['text'].str.split(' ', 1).apply(pd.Series)
df
```
